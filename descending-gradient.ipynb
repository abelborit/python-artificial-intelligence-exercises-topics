# Algoritmo de Gradiente Descendente

El gradiente descendente es un algoritmo de optimización utilizado para encontrar el mínimo de una función. Es ampliamente utilizado en el campo del aprendizaje automático y la inteligencia artificial para entrenar modelos de machine learning. 

El gradiente de una función es un vector que contiene las derivadas parciales de la función con respecto a cada una de sus variables independientes. Entonces, el gradiente es esencialmente la derivada de una función multivariable y nos proporciona información importante sobre la pendiente y la dirección del cambio de la función en un punto dado. La dirección del gradiente nos dice en qué dirección debemos ajustar el valor de x para minimizar la función de pérdida, y la magnitud del gradiente nos indica qué tan pronunciado es ese cambio.

El concepto básico del gradiente descendente es ajustar iterativamente los parámetros de un modelo en la dirección opuesta al gradiente de la función de pérdida. Esto implica moverse hacia abajo en la función de pérdida en pequeños pasos (determinados por la tasa de aprendizaje) hasta alcanzar un mínimo local o global.

El proceso de gradiente descendente se puede resumir en los siguientes pasos:

1. **Inicialización de parámetros:** Seleccionamos valores iniciales para los parámetros del modelo. Estos valores pueden ser aleatorios o elegidos de alguna otra manera.

2. **Cálculo del gradiente:** Calculamos el gradiente de la función de pérdida con respecto a cada parámetro del modelo. El gradiente indica la dirección y la magnitud del cambio más pronunciado en la función de pérdida.

3. **Actualización de parámetros:** Actualizamos los parámetros del modelo moviéndolos en la dirección opuesta al gradiente, multiplicado por la tasa de aprendizaje. Esto se hace para minimizar la función de pérdida.

4. **Convergencia:** Repetimos los pasos 2 y 3 hasta que se cumpla algún criterio de convergencia, como alcanzar un número máximo de iteraciones o una mejora mínima en la función de pérdida.

El gradiente descendente puede tener diferentes variantes, como el gradiente descendente estocástico (SGD), el mini-batch gradient descent, y el gradient descent con momentum, entre otros. Estas variantes pueden ser útiles para acelerar la convergencia o evitar mínimos locales.

En resumen, el gradiente descendente es un algoritmo fundamental en el aprendizaje automático que permite entrenar modelos ajustando iterativamente sus parámetros para minimizar la función de pérdida.


```python
# # # EJEMPLO BÁSICO
# # Supongamos que queremos minimizar la función de pérdida cuadrática, dada por ( f(x) = x^2 ) donde la gradiente de esa función sería la derivada parcial de esa función que sería 2x es decir, ( gradient = 2 * x ). 

# # # Inicialización de variables
# x = 3  # Valor inicial de x
# alpha = 0.1  # Tasa de aprendizaje
# num_iterations = 25  # Número de iteraciones

# # # Algoritmo de gradiente descendente
# # # Este código imprimirá los valores de x en cada iteración del algoritmo de gradiente descendente. Se observará cómo x se acerca al valor óptimo de 0, que minimiza la función ( f(x) = x^2 ).
# for i in range(num_iterations):
#     gradient = 2 * x  # Gradiente de la función
#     x = x - alpha * gradient  # Actualización de x utilizando el gradiente descendente
#     print("Iteración:", i + 1, " - x:", x)
```

```python
# Para graficar el comportamiento del algoritmo de gradiente descendente, podemos trazar la función de pérdida en función del parámetro que estamos optimizando (en este caso sería para x). Esto nos permitirá visualizar cómo el algoritmo de gradiente descendente busca el mínimo de la función de pérdida a medida que actualiza iterativamente el valor de x.
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de pérdida
def loss_function(x):
    return x ** 2

# Inicialización de variables
x = 3  # Valor inicial de x
alpha = 0.1  # Tasa de aprendizaje
num_iterations = 10  # Número de iteraciones

# Listas para almacenar valores de x y de la función de pérdida en cada iteración
x_values = [x]
loss_values = [loss_function(x)]

print("Initial Values")
print("x_values", x_values)
print("loss_values", loss_values, "\n")

# Algoritmo de gradiente descendente
for i in range(num_iterations):
    gradient = 2 * x  # Gradiente de la función
    x = x - alpha * gradient  # Actualización de x utilizando el gradiente descendente
    # print("Iteración:", i + 1, " - x:", x)
    x_values.append(x)
    loss_values.append(loss_function(x))

print("All Values Iterations")
print("x_values", x_values)
print("loss_values", loss_values)

# Gráfico de la función de pérdida y el comportamiento de x. Esto permitirá visualizar cómo el algoritmo de gradiente descendente converge hacia el mínimo de la función de pérdida a medida que se realizan más iteraciones.
plt.figure(figsize=(10, 5))

# Función de pérdida
plt.subplot(1, 2, 1) # se utiliza para crear subgráficos dentro de una figura. La sintaxis general de esta función es plt.subplot(nrows, ncols, index), donde nrows es el número de filas de subgráficos, ncols es el número de columnas de subgráficos, y index es el índice del subgráfico actual.
plt.plot(np.arange(-4, 4, 0.1), loss_function(np.arange(-4, 4, 0.1)), color='blue')
plt.title('Función de Pérdida')
plt.xlabel('x')
plt.ylabel('Pérdida')
plt.grid(True)

# Comportamiento de x durante las iteraciones
plt.subplot(1, 2, 2)
plt.plot(range(num_iterations + 1), x_values, marker='o', color='red')
plt.title('Comportamiento de x')
plt.xlabel('Iteración')
plt.ylabel('x')
plt.grid(True)

plt.tight_layout() # ajusta automáticamente los parámetros de diseño de la figura para que los elementos (subgráficos, etiquetas, títulos, etc.) se ajusten correctamente dentro del área de la figura sin superponerse. Esto es útil cuando se trabaja con múltiples subgráficos en una misma figura, ya que ayuda a mejorar la legibilidad y la apariencia general del gráfico.
plt.show()
```
