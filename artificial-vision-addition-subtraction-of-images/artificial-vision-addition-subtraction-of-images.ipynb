# Suma y Resta de Imágenes
La aritmética de imágenes es la implementación de operaciones aritméticas estándar, como suma, resta, multiplicación y división, en imágenes. La aritmética de imágenes tiene muchos usos en el procesamiento de imágenes, tanto como paso preliminar en operaciones más complejas como por sí sola.

#### Suma de imágenes:
La suma de imágenes se realiza tomando cada píxel de la primera imagen y sumándole el valor correspondiente del píxel en la segunda imagen. Si la suma de los valores de los píxeles excede el valor máximo (generalmente 255 en una imagen en escala de grises), se limita a 255 para evitar el desbordamiento. Por ejemplo, si un píxel en la primera imagen tiene un valor de 200 y el píxel correspondiente en la segunda imagen tiene un valor de 100, la suma resultante para ese píxel sería 255.

#### Resta de imágenes:
La resta de imágenes se realiza tomando cada píxel de la primera imagen y restando el valor correspondiente del píxel en la segunda imagen. Si el resultado de la resta es negativo, se limita a cero para evitar valores negativos en la imagen resultante, ya que los valores de píxeles en imágenes generalmente están en el rango de 0 a 255. Por ejemplo, si un píxel en la primera imagen tiene un valor de 200 y el píxel correspondiente en la segunda imagen tiene un valor de 100, la resta resultante para ese píxel sería 100.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def sumar_imagenes(imagen1, imagen2):
    # Asegurarse de que las dos imágenes tengan el mismo tamaño
    if imagen1.shape != imagen2.shape:
        raise ValueError("Las imágenes deben tener el mismo tamaño")

    # Crear una nueva matriz para el resultado
    resultado = imagen1.copy()  # Crear una copia de la primera imagen
    for i in range(resultado.shape[0]):
        for j in range(resultado.shape[1]):
            # Sumar los valores de píxeles y asegurarse de que no se excedan de 255
            resultado[i][j] = min(resultado[i][j] + imagen2[i][j], 255)
    
    return resultado

def restar_imagenes(imagen1, imagen2):
    # Asegurarse de que las dos imágenes tengan el mismo tamaño
    if imagen1.shape != imagen2.shape:
        raise ValueError("Las imágenes deben tener el mismo tamaño")

    # Crear una nueva matriz para el resultado
    resultado = imagen1.copy()  # Crear una copia de la primera imagen
    for i in range(resultado.shape[0]):
        for j in range(resultado.shape[1]):
            # Restar los valores de píxeles y asegurarse de que no sean negativos
            resultado[i][j] = max(resultado[i][j] - imagen2[i][j], 0)
    
    return resultado

# Cargar las imágenes en escala de grises con OpenCV
imagen1 = cv2.imread('./image3.jpg', cv2.IMREAD_GRAYSCALE)
imagen2 = cv2.imread('./image4.jpg', cv2.IMREAD_GRAYSCALE)

# Realizar las operaciones de suma y resta con las funciones creadas
resultado_suma = sumar_imagenes(imagen1, imagen2)
resultado_resta = restar_imagenes(imagen1, imagen2)

# ******************************************************************************************************************************
# # Mostrar los resultados en una ventana aparte
# cv2.imshow('Suma de imágenes', resultado_suma)
# cv2.imshow('Resta de imágenes', resultado_resta)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# ******************************************************************************************************************************

# mostrar la imagen original 1 y la imagen original 2 distribuidas horizontalmente
# mostrar la suma y resta de las imágenes distribuidas horizontalmente
plt.figure(figsize=(14, 7))

# mostrar la imagen original 1
plt.subplot(2, 2, 1)
plt.imshow(imagen1, cmap='gray')
plt.title('Imagen 1')

# mostrar la imagen original 2
plt.subplot(2, 2, 2)
plt.imshow(imagen2, cmap='gray')
plt.title('Imagen 2')

# mostrar la suma de imágenes
plt.subplot(2, 2, 3)
plt.imshow(resultado_suma, cmap='gray')
plt.title('Suma de imágenes')

# mostrar la resta de imágenes
plt.subplot(2, 2, 4)
plt.imshow(resultado_resta, cmap='gray')
plt.title('Resta de imágenes')

# Ajustar y mostrar el gráfico
plt.subplots_adjust(hspace=0.4, wspace=0.015)
plt.show()
```
