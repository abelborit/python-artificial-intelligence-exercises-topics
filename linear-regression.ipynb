# Regresion Lineal
La regresión lineal es un método estadístico utilizado para modelar la relación entre una variable dependiente (o respuesta) y una o más variables independientes (o características). Es uno de los algoritmos más simples y ampliamente utilizados en análisis de datos y aprendizaje automático.

En el caso más simple, llamado regresión lineal simple, se modela la relación entre dos variables: una variable independiente (X) y una variable dependiente (y). Se asume que la relación entre estas dos variables puede aproximarse mediante una línea recta. Esta línea recta se utiliza para predecir los valores de la variable dependiente (y) en función de los valores de la variable independiente (X).

La ecuación de una línea recta en su forma más simple es:

\[ y = mx + b \]

Donde:
- \( y \) es la variable dependiente (o respuesta) que queremos predecir.
- \( x \) es la variable independiente (o características) que se utiliza para hacer la predicción.
- \( m \) es la pendiente de la línea, que representa el cambio en \( y \) por cada unidad de cambio en \( x \).
- \( b \) es la intersección con el eje \( y \), es decir, el valor de \( y \) cuando \( x \) es igual a cero.

El objetivo de la regresión lineal es encontrar los mejores valores para los coeficientes \( m \) y \( b \) que minimicen la diferencia entre los valores predichos (\( mx + b \)) y los valores reales (\( y \)). Esto se hace utilizando métodos como el método de mínimos cuadrados, que calcula los coeficientes de manera que la suma de los cuadrados de las diferencias entre los valores predichos y los valores reales sea mínima.

En resumen, la regresión lineal es un enfoque para modelar la relación entre variables continuas mediante una línea recta, y se utiliza para hacer predicciones sobre la variable dependiente en función de la variable independiente.

```python
# Primero, importamos las librerías necesarias. En este caso, necesitaremos numpy para trabajar con arrays multidimensionales y realizar operaciones matemáticas eficientes trabajando con operaciones básicas y matplotlib para realizar las gráficas necesarias.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ahora generamos algunos datos ficticios para probar nuestra función de regresión lineal.
# Aquí creamos dos arrays NumPy, X y y, que representan nuestros datos de entrada (características) y salida (etiquetas) respectivamente. Estos son solo datos ficticios para propósitos de demostración.

# Datos de entrada (características)
X = np.array([1, 2, 3, 4, 5])

# Datos de salida (etiquetas)
y = np.array([2, 3, 4, 5, 6])
```

```python
# Esta función toma dos argumentos: X, que es un array NumPy de características, y y, que es un array NumPy de etiquetas. Calcula los coeficientes de la regresión lineal utilizando el método de mínimos cuadrados.
def linear_regression(X, y):
  print("X:", X)
  print("y:", y)

  # Calculamos la media de X e y. La razón por la cual calculamos la media o promedio es para encontrar el valor promedio de las variables de entrada (X) y de salida (y). Esto es importante porque estamos interesados en calcular la pendiente (slope) y la intersección (intercept) de la línea de regresión que mejor se ajuste a nuestros datos. 
  x_mean = np.mean(X)
  y_mean = np.mean(y)
  print("x_mean:", x_mean)
  print("y_mean:", y_mean)
  # ************************************************************************************************************************

  # Calculamos los coeficientes de la regresión lineal mediante el método de mínimos cuadrados (numerator y denominator calculan partes de la fórmula para el coeficiente de pendiente (slope))
  # En conjunto, estas dos partes son fundamentales para calcular la pendiente (slope) de la línea de regresión en el método de mínimos cuadrados. La pendiente se calcula dividiendo el numerador por el denominador. Esta fórmula nos ayuda a encontrar la pendiente de la línea que mejor se ajusta a nuestros datos, minimizando la suma de los cuadrados de las diferencias entre los valores predichos y los valores reales.

  # (X - x_mean) y (y - y_mean) calculan las desviaciones de cada valor de X e y respecto a sus medias respectivas. Esto nos da una medida de cuánto cada punto de datos se desvía de la media. Luego, multiplicamos estas desviaciones para cada par de datos. Finalmente, sumamos todas estas multiplicaciones para obtener el numerador.
  numerator = np.sum((X - x_mean) * (y - y_mean))

  # (X - x_mean) ** 2 calcula el cuadrado de las desviaciones de cada valor de X respecto a su media. Esto nos proporciona una medida de dispersión de los valores de X. Luego, sumamos todos estos cuadrados. Esto nos da el denominador.
  denominator = np.sum((X - x_mean) ** 2)
  print("numerator:", numerator)
  print("denominator:", denominator)
  # ************************************************************************************************************************

  # slope (pendiente) y intercept (intersección) son los coeficientes de la recta de regresión que devuelve la función.

  # Para la pendiente (slope), necesitamos calcular la covarianza entre X e y, así como la varianza de X. La covarianza nos indica cómo varían dos variables juntas, mientras que la varianza nos indica cuánto varía una sola variable. Dividiendo la covarianza entre la varianza de X, obtenemos la pendiente de la línea de regresión.
  slope = numerator / denominator

  # Para la intersección (intercept), necesitamos la media de y y la pendiente calculada previamente. La intersección es el punto en el que la línea de regresión corta el eje y cuando X es igual a cero.
  intercept = y_mean - slope * x_mean
  print("slope:", slope)
  print("intercept:", intercept)

  # ************************************************************************************************************************
  return slope, intercept
```

```python
# Creamos una función para predecir los valores de salida (y) dados los valores de entrada (X) y los coeficientes calculados.
# Esta función toma un valor de entrada X, así como los coeficientes de la regresión lineal slope e intercept, y devuelve la predicción correspondiente usando la ecuación de la recta (y = mx + b).
def predict(X, slope, intercept):
    return slope * X + intercept
```

```python
# Aplicamos nuestra función de regresión lineal a los datos.
# Llamamos a la función linear_regression pasando nuestros datos X e y como argumentos. Esto nos devuelve los coeficientes de la regresión lineal (slope y intercept), que almacenamos en variables.
slope, intercept = linear_regression(X, y)
```

```python
# # Finalmente, podemos hacer predicciones utilizando los coeficientes calculados.
# # Aquí, especificamos un nuevo valor de entrada new_X para el cual queremos hacer una predicción. Llamamos a la función predict pasando el nuevo valor de entrada new_X y los coeficientes de la regresión lineal (slope e intercept). La función devuelve la predicción correspondiente, que imprimimos en pantalla.

# # Predicción para un nuevo valor de entrada
new_X = 6
new_X = 8
prediction = predict(new_X, slope, intercept)
print("La predicción para X =", new_X, "es el valor de y =", prediction)

# # Gráfico de los datos originales y la regresión lineal para un nuevo valor
plt.scatter(X, y, color='blue', label='Datos originales')
plt.plot(X, slope * X + intercept, color='red', label='Regresión lineal') # se utiliza para trazar líneas o curvas conectando puntos dados por las coordenadas x y y
plt.scatter(new_X, prediction, color='green', label='Predicción para X=6') # se utiliza para crear un gráfico de dispersión, también conocido como diagrama de dispersión o gráfico de puntos. Este tipo de gráfico se utiliza para visualizar la relación entre dos variables representando puntos en un plano cartesiano, donde cada punto representa una observación en los datos. Se utiliza para trazar puntos individuales sin conectarlos por líneas.
plt.xlabel('variable independiente (X)')
plt.ylabel('variable dependiente (y)')
plt.title('Regresión Lineal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Finalmente, podemos hacer predicciones utilizando los coeficientes calculados.
# Aquí, especificamos nuevos valores de entrada en new_X_values para el cual queremos hacer las predicciones. Llamamos a la función predict pasando los nuevos valores de entrada new_X_values y los coeficientes de la regresión lineal (slope e intercept). La función devuelve las predicciones correspondientes, que imprimimos en pantalla.

# Predicciones para nuevos valores de entrada
new_X_values = np.array([6, 7, 8, 9, 10])
predictions = predict(new_X_values, slope, intercept)
print("La predicción para X =", new_X_values, "es el valor de y =", predictions)

# Gráfico de los datos originales y la regresión lineal para nuevos valores
plt.scatter(X, y, color='blue', label='Datos originales')
plt.plot(X, slope * X + intercept, color='red', label='Regresión lineal') # se utiliza para trazar líneas o curvas conectando puntos dados por las coordenadas x y y
plt.scatter(new_X_values, predictions, color='green', label='Predicción para X=6') # se utiliza para crear un gráfico de dispersión, también conocido como diagrama de dispersión o gráfico de puntos. Este tipo de gráfico se utiliza para visualizar la relación entre dos variables representando puntos en un plano cartesiano, donde cada punto representa una observación en los datos. Se utiliza para trazar puntos individuales sin conectarlos por líneas.
plt.xlabel('variable independiente (X)')
plt.ylabel('variable dependiente (y)')
plt.title('Regresión Lineal')
plt.legend()
plt.grid(True)
plt.show()
```
